```python
import matplotlib.pyplot as plt
import matplotlib_venn as venn

S = {1, 2, 3}
T = {0, 2, -1, 5}

venn.venn2([S, T], set_labels = ('S','T'))
plt.show()
```

```python
S = {1, 2, 3, 4, 5}
T = {0, 2, -1, 5, 10}
U = {1, 2, 3}
venn.venn3([S, T, U], set_labels = ('S','T', 'U'))
plt.show()
```
